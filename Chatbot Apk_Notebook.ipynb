## TAHAP 1: RETRIEVER

Retriever adalah komponen utama dalam sistem **Retrieval-Augmented Generation (RAG)** yang bertugas untuk mengambil dokumen atau informasi relevan dari vectorstore berdasarkan query pengguna.

Retriever bertujuan untuk:
- Mempersempit ruang pencarian dokumen agar sesuai dengan kebutuhan query pengguna.
- Memberikan daftar dokumen relevan yang mendukung jawaban yang lebih kontekstual dan akurat.

Fase dalam Retriever
- Loading Document
- Spliting Document
- Embedding Document
- Vectorstore

### Load Document

Langkah-langkah proses loading dokumen:
1. **Mendefinisikan file**: File yang akan digunakan didefinisikan dalam variabel `file_paths`.

2. **Menggunakan PyPDFLoader dan Load**:
  - PyPDFLoader adalah method bawaan LangChain untuk membaca file PDF.
  - Load adalah method untuk memuat file pdf dan mengubahnya menjadi format dokumen LangChain.

3. **Menambahkan Metadata**: Metadata adalah informasi tambahan yang memberikan deskripsi tentang dokumen, seperti:
  - Sumber dokumen: nama file (source).
  - Halaman dokumen: nomor halaman (page).

4. **Gabungkan Dokumen**: Semua dokumen dari berbagai file digabungkan ke dalam satu list `all_documents`.

5. **Verifikasi Dokumen**: Memeriksa jumlah total dokumen yang berhasil dimuat.


Dokumen eksternal yang digunakan:
- Buku Panduan Strategi Nasional Percepatan Pencegahan Stunting
- Peraturan Menteri Kesehatan Republik Indonesia Nomor 2 Tahun 2020 tentang
Standar Antropometri Anak.


```python
from langchain_community.document_loaders import PyPDFLoader

# Mencari dokumen (file PDF) yang akan dimuat
file_paths = [
    "/content/drive/MyDrive/IF-46-07/Big Data Laboratory/Chatbot Gizi dan Stunting/docs/1-Strategi Nasional Percepatan Pencegahan Stunting.pdf",
    "/content/drive/MyDrive/IF-46-07/Big Data Laboratory/Chatbot Gizi dan Stunting/docs/2-Peraturan Menteri Kesehatan Republik Indonesia Nomor 2 Tahun 2020 tentang Standar Antropometri Anak.pdf",
    "/content/drive/MyDrive/IF-46-07/Big Data Laboratory/Chatbot Gizi dan Stunting/docs/3-Stunting di Indonesia dan Faktor Determinan.pdf",
    "/content/drive/MyDrive/IF-46-07/Big Data Laboratory/Chatbot Gizi dan Stunting/docs/4-Buku Stunting.pdf"
]

# List untuk menyimpan semua dokumen
all_documents = []

# Load setiap file PDF
for path in file_paths:
  loader = PyPDFLoader(path) # Inisialisasi PyPDFLoader untuk membaca file
  documents = loader.load()  # Memuat dokumen dan mengubah ke format LangChain

  # Menambahkan metadata ke setiap dokumen
  for idx, doc in enumerate(documents):
    doc.metadata["source"] = path.split("/")[-1]  # Tambahkan nama file
    doc.metadata["page"] = idx + 1  # Tambahkan nomor halaman (mulai dari 1)

  # Gabungkan dokumen hasil load ke dalam list utama
  all_documents.extend(documents)

# Verifikasi jumlah dokumen yang dimuat
print(f"Total documents loaded: {len(all_documents)}")
```

```python
# Menampilkan metadata setiap dokumen, termasuk informasi halaman
for doc in all_documents:
  print(doc.metadata)
```

Menampilkan contoh konten file pdf dari dokumen pada halaman tertentu

```python
from IPython.display import Markdown

# Fungsi untuk menampilkan konten satu halaman dari dokumen tertentu
def display_single_page_content(documents, page_number, document_name=None):
    # Filter dokumen berdasarkan nomor halaman
    filtered_docs = [
        doc for doc in documents
        if doc.metadata.get("page") == page_number and (document_name is None or doc.metadata.get("source") == document_name)
    ]

    if not filtered_docs:
        print(f"Tidak ditemukan dokumen pada halaman {page_number} dengan nama file {document_name or 'any document'}.")
        return

    # Ambil dokumen pertama yang cocok
    selected_doc = filtered_docs[0]

    # Tampilkan isi halaman
    display(Markdown(f"""### **Document from Page [{page_number}]**\n\n**Source Document**:\n\n{selected_doc.metadata['source']}\n\n**Page Content**:\n\n{selected_doc.page_content}\n\n**Metadata:**\n\n{selected_doc.metadata}"""))

# Menampilkan isi konten buku stunting pada halaman 28
document_name = "4-Buku Stunting.pdf"
display_single_page_content(all_documents, page_number=28, document_name=document_name)
```

### Split Document

Proses **splitting** adalah langkah kedua dalam pipeline Retriever. Langkah ini bertujuan untuk memecah dokumen besar menjadi potongan-potongan teks kecil yang disebut chunks. Chunking mempermudah pengolahan data oleh model LLM karena menjaga konteks tetap terfokus pada unit teks yang lebih kecil.

#### RecursiveCharacterTextSplitter
RecursiveCharacterTextSplitter adalah salah satu metode splitting dalam LangChain yang membagi teks berdasarkan panjang karakter tertentu, dengan mempertahankan konteks sebanyak mungkin.

#### Splitting Parameters

**chunk_size**
- Ukuran maksimal panjang karakter dalam satu chunk
- Contoh: Jika teks memiliki 10.000 karakter dan chunk_size=1000, teks akan dibagi menjadi 10 chunk.

**chunk_overlap**
- Banyaknya karakter yang diambil dari chunk sebelumnya dan akan ikut dimuat pada chunk berikutnya
- Contoh: Jika teks dipecah menjadi chunk A dan B, overlap memastikan akhir chunk A mirip dengan awal chunk B.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Serring parameter untuk RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""],
    chunk_size=1600,
    chunk_overlap=200
)

# Proses text splitting
chunks = text_splitter.split_documents(all_documents)

# Menampilkan jumlah chunk yang dihasilkan dari text splitting
print(f"Number of chunks: {len(chunks)}")
```

```python
# Menampilkan contoh chunk
print(f"{chunks[500].page_content}")
print(f"Metadata: {chunks[500].metadata}")
```

### Embedd Document

**Embedding** adalah proses mengubah teks menjadi representasi numerik (vektor).

Dokumen yang sudah dalam bentuk vektor akan mempermudah sistem dalam proses pencarian kesamaan (similarity search), yaitu proses pencarian dokumen yang relevan dengan query pengguna.

Dokumen yang memiliki tingkat kesamaan paling tinggi akan dipilih dan akan diproses lebih lanjut menjadi prompt dan diserahkan kepada LLM untuk memberikan jawaban.

#### Create Embedding

### Retriever

Setelah vectorstore berhasil dibuat artinya **RETRIEVER** telah siap untuk digunakan.

**Retriever** bertanggung jawab untuk mencari dan mengembalikan daftar dokumen yang relevan dengan query pengguna dari vectorstore.

Retriever dalam Langhain memiliki parameter yang harus ditentukan diawal. Parameter ini akan menentukan bagaimana proses pencarian dokumen dilakukan.

Tiga parameter dalam retriever:
- Search Type
- Search Kwargs


Jenis-jenis search type dalam retriever :
- similarity search (deafult)
- max marginal relevance search
- similarity score threshold

Jenis-jenis search kwargs dalam retriever:
- Nilai k
- score_trashold


Kita dapat mengkombinasikan` search_type` dengan `search_kwargs` untuk mendapatkan konfigurasi terbaik, agar retriever dapat menjalankan tugas nya dengan baik dalam mencari dokumen yang paling relevan dengan query pengguna

```python
# Query percobaan untuk menguji bagaimana retriever mengembalikan dokumen relevan dari vectorstore.
query = "Dasar hukum dari pemerintah tentang gizi dan stunting"
```

#### Similarity Search

Mengembalikan dokumen berdasarkan tingkat kesamaan tertinggi dengan query.


```python
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 20})
results = retriever.get_relevant_documents(query)
for i, doc in enumerate(results):
    print(f"Document {i+1}:\nContent: {doc.page_content}\nMetadata: {doc.metadata}\n")
```

#### Max Marginal Relevance Search (MMR)

MMR akan mengembalikan dokumen dengan kesamaan/relevansi yang tinggi dan memperhatikan keberagaman (diverse) dokumen, sehingga dapat menghindari adanya duplikasi informasi dari dokumen yang diperoleh

```python
retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={"k": 5})
results = retriever.get_relevant_documents(query)
for i, doc in enumerate(results):
    print(f"Document {i+1}:\nContent: {doc.page_content}\nMetadata: {doc.metadata}\n")
```

#### Similarity Score Threshold
Mengembalikan dokumen yang memiliki skor relevansi di atas ambang batas atau treshold yang telah ditetapkan.


```python
retriever = vectorstore.as_retriever(search_type="similarity_score_threshold", search_kwargs={"score_threshold": 0.75, "k": 5})
results = retriever.get_relevant_documents(query)
for i, doc in enumerate(results):
    print(f"Document {i+1}:\nContent: {doc.page_content}\nMetadata: {doc.metadata}\n")
```

#### Contextual Compression Retriever

Ide: Dokumen relevan yang dikembalikan oleh retriever sering kali tidak benar-benar relevan karena masih ada beberapa informasi yang kurang relevan.

Contextual Compression Retriever dilakukan agar retriever benar-benar mengembalikan dokumen yang relevan, dengan menghapus atau menghilangkan dokumen yang tidak relevan

Tujuan:
- Memastikan dokumen yang diambil oleh retriever benar-benar relevan dengan query.
- Mengurangi dokumen yang tidak relevan untuk mengurangi biaya token yang terbuang.
- Meningkatkan akurasi respons model dengan memberikan dokumen yang lebih terfokus.

Proses Contextual Compression Retriever:
1. Mengambil daftar dokumen relevan hasil dari Retriever
2. Mengolah dokumen relevan tersebut untuk menghilangkan teks yang tidak relevan dan mempertahankan teks yang relevan:
    - Splitting dokumen hasil retriever menjadi potongan-potongan kecil (chunk) untuk evaluasi lebih mendetail.
    - Removing redundant chunk
    - Filtering untuk mendapatkan chunk yang sangat relevan dengan query.
    - Reordering dengan mengurutkan ulang dokumen dengan relevansi tertinggi di bagian atas untuk meningkatkan performa LLM.

```python
# Helper function untuk menampilkan hasil dokumen hasil pencarian pada vectorstore agar lebih rapih
def pretty_print_docs_with_metadata(docs):
    """
    Helper function to display documents with their metadata in a readable format.
    """
    print(
        f"\n{'-' * 100}\n".join(
            [
                f"Document {i+1}:\n\nPage Content:\n{doc.page_content}\n\nMetadata:\n{doc.metadata}\n"
                for i, doc in enumerate(docs)
            ]
        )
    )
```

```python
# Import library yang diperlukan
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.document_transformers import EmbeddingsRedundantFilter,LongContextReorder
from langchain.retrievers.document_compressors import EmbeddingsFilter
from langchain.retrievers import ContextualCompressionRetriever
```

```python
# Base retriever menggunakan similarity search
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 20})

# Inisialisasi Embeddings
embedding_model = OpenAIEmbeddings(model="text-embedding-ada-002")

# Splitting menggunakan CharacterTextSplitter untuk memecah dokumen menjadi chunk kecil
splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=0)

# Removing dengan menghapus chunk yang redundant
redundant_filter = EmbeddingsRedundantFilter(embeddings=embedding_model)

# Filtering dengan memilih chunk yang sangat relevan dengan query
relevant_filter = EmbeddingsFilter(
  embeddings=embedding_model,
  similarity_threshold=0.75,  # Ambang batas relevansi
  k=16,  # Jumlah maksimal chunk yang relevan
)

# Reordering menggunakan LongContextReorder
# Dokumen yang lebih relevan akan ditempatkan di awal dan akhir daftar dokumen
# Dokumen yang dianggap kurang relevan akan ditempatkan di bagian tengah daftar dokumen
reordering = LongContextReorder(embeddings=embedding_model)

# Document Compressor Pipeline
pipeline_compressor = DocumentCompressorPipeline(
    transformers=[splitter, redundant_filter, relevant_filter, reordering]
)

# Membuat Contextual Compression Retriever
compression_retriever = ContextualCompressionRetriever(
    base_compressor=pipeline_compressor,
    base_retriever=retriever
)
```

```python
# Query percobaan
query = "Dasar hukum dari pemerintah tentang gizi dan stunting"
```

```python
# Menggunakan Contextual Compression Retriever untuk mendapatkan dokumen relevan
compressed_docs = compression_retriever.get_relevant_documents(query)

# Menampilkan hasil dengan helper function
pretty_print_docs_with_metadata(compressed_docs)
```

### Testing Query

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
```

```python
# Helper function untuk menggabungkan hasil dokumen yang relevan menjadi string
def docs2str(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
# Membuat prompt
template = """Jawab pertanyaan berikut berdasarkan hanya pada konteks di bawah ini:
{context}

Pertanyaan: {question}

Jawaban:"""
prompt = ChatPromptTemplate.from_template(template)
```

```python
# Inisialisasi model LLM

# Temperature = 0 -> model memberikan jawaban yang berdasarkan fakta dari dokumen relevan
# Semakin kecil nilai temperature maka kreatifitas LLMs semakin rendah, begitu juga sebaliknya
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
# Membuat pipeline (RAG chain) dengan mengintegrasikan retriever, prompt, dan LLM.
rag_chain = (
    {"context": compression_retriever | docs2str, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
# Membuat query pertanyaan pengguna
query = "Apa dasar hukum pemerintah terkait gizi dan stunting?"
```

```python
# Menampilkan dokumen yang relevan dari hasil compression_retriever
# Dokumen ini yang akan dijadikan konteks atau pengetahuan bagi LLMs untuk menjawab pertanyaan

# Mengambil dokumen relevan dari retriever
relevant_docs = compression_retriever.get_relevant_documents(query)

# Menampilkan dokumen relevan
print("Dokumen Relevan:")
pretty_print_docs_with_metadata(relevant_docs)
```

```python
# Menampilkan jawaban dari LLMs

# Menjalankan RAG Chain
response = rag_chain.invoke(query)

# Menampilkan jawaban dari LLM
print(f"Jawaban LLM:\n{response}")
```

### Stand-alone Question

**Standalone Question** adalah pertanyaan yang tidak mengandung kata ganti atau kalimat lengkap dari sebuah pertanyaan.

Standalone question diperoleh dari **follow-up question** atau query pengguna yang menggunakan kata ganti, dalam bahasa indonesia kata ganti dapat berupa "ini, itu, tersebut", atau dalam bahasa inggris contoh kata ganti adalah "it, that, this"

Jadi untuk dapat mengetahui konteks dari follow-up question, dibutuhkan adanya **chat history** atau riwayat percakapan dari pertanyaan-pertanyaan dan jawaban-jawaban yang telah dilakukan sebelumnya di dalam chatbot.

Informasi terkait query pengguna yang mengandung follow-up question akan di proses oleh LLMs bersama dengan chat history untuk menghasilkan standalone question.

```python
from langchain_core.messages import HumanMessage, AIMessage
# Menyimpan percakapan awal ke dalam chat history
chat_history = []
chat_history.extend([
    HumanMessage(content=query),
    AIMessage(content=response)
])
```

```python
# Melihat daftar percakapan dalam chat history
chat_history
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# Membuat prompt untuk menghasilkan standalone question
contextualize_q_system_prompt = (
    "Diberikan riwayat percakapan (chat history) dan pertanyaan. "
    "Pertanyaan mungkin merujuk pada konteks yang ada di dalam chat history, "
    "Susun ulang pertanyaan menjadi standalone question atau kalimat pertanyaan yang lengkap."
    "Penyusunan ulang pertanyaan harus dapat dimengerti dan tanpa tambahan dari chat history."
    "Jangan berikan jawaban, anda hanya diminta untuk menyusun pertanyaan yang mungkin merujuk "
    "pada konteks di dalam chat history."
)

contextualize_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", contextualize_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)

# Membuat contextualize_chain
contextualize_chain = contextualize_q_prompt | llm | StrOutputParser()
```

```python
# Follow-up question dari pengguna
follow_up_question = "Bagaimana strategi pencapaiannya terkait hal itu?"

# Menghasilkan standalone question berdasarkan chat history dan follow-up question
standalone_question = contextualize_chain.invoke({
    "input": follow_up_question,
    "chat_history": chat_history
})

# Menampilkan hasil
print("Standalone Question:")
print(standalone_question)
```

### CreateHistoryAwareRetriever

`create_history_aware_retriever` adalah library dalam LangChain yang dirancang untuk menggabungkan langkah mengubah **follow-up question** menjadi **standalone question** dan melakukan pencarian dokumen yang relevan dalam satu proses.

```python
from langchain.chains import create_history_aware_retriever
# Membuat history-aware retriever
history_aware_retriever = create_history_aware_retriever(
    llm,                 # Model LLM yang akan digunakan untuk standalone question
    compression_retriever,           # Base retriever, seperti compression_retriever
    contextualize_q_prompt  # Prompt untuk mengenerate standalone question
)
```

```python
# Menjalankan pencarian dokumen relevan menggunakan follow-up question
relevant_docs = history_aware_retriever.invoke({
    "input": "Bagaimana strategi pencapaiannya terkait hal itu?",  # Follow-up question
    "chat_history": chat_history            # Chat history untuk konteks
})

# Menampilkan dokumen relevan
print("Dokumen Relevan:")
pretty_print_docs_with_metadata(relevant_docs)
```

### Generating Answer

 Mengetes keseluruhan prosess yang telah dibuat dengan memberikan query pertanyaan berupa follow-up question kepada LLMs agar dijawab.

Proses ini melibatkan:
1. Penyusunan standalone question dari follow-up question yang diberikan dan percakapan yang sudah ada dalam chat history.
2. Pencarian dokumen berdasarkan standalone question
3. Menyiapkan prompt untuk diberikan kepada LLMs
4. Memberikan prompt dan hasil dokumen relevan kepada LLMs untuk menjawab pertanyaan.

```python
chat_history
```

```python
# Menyiapkant prompt
qa_prompt = ChatPromptTemplate.from_messages([
    ("system", "Anda adalah asisten AI yang membantu memberikan jawaban berdasarkan dokumen relevan berikut."),
    ("system", "Konteks: {context}"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])
```

```python
# Membuat chain antara prompt yang telah dibuat dengan llm
from langchain.chains.combine_documents import create_stuff_documents_chain
question_answer_chain = create_stuff_documents_chain(llm, qa_prompt)
```

```python
# Membuat chain antara retriever dengan question_answer_chain yang telah dibuat di atas
from langchain.chains import create_retrieval_chain
rag_chain_full = create_retrieval_chain(history_aware_retriever, question_answer_chain)
```

```python
follow_up_question = "Bagaimana strategi pencapaiannya terkait hal itu?"
```

```python
response_follow_up = rag_chain_full.invoke({
    "input": follow_up_question,
    "chat_history": chat_history  # Menggunakan chat history yang telah ada
})
```

```python
# Menampilkan jawaban
response_content = response_follow_up.get("answer", "Jawaban tidak ditemukan.")
print(f"Jawaban LLM dari Follow-Up Question:\n{response_content}")
```
